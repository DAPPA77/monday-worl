```python
def binary(list,target):
    first=0
    last=len(list)-1
    while first<=last:
        mid=(first+last)//2
        if list[mid]==target:
            return mid
        elif list[mid]<target:
            first=mid+1
        else:
            last=mid-1
    return None
number=[]
for b in range(11):
    number.append(b)
       
```

```python
binary(number,6)
```

```python
def  recursive(list,target):
    if len(list)==0:
        return False
    else:
        mid=len(list)//2
        if list[mid]==target:
            return True
        else:
            if list[mid]<target:
                return recursive(list[mid+1:],target)
            else:
                return recursive(list[:mid],target)
number=[]
for b in range(11):
    number.append(b)
```

```python
recursive(number,6)
```
